```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
titanic = pd.read_csv("titanic.csv")
titanic
```

```python
titanic.head(5)
```

```python
# titanic = titanic.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
titanic.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)
titanic
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(titanic.drop(columns=["Survived"]), titanic["Survived"], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
titanic.isnull().sum()
```

# Applying Imputation

```python
si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy="most_frequent")

X_train_age = si_age.fit_transform(X_train[["Age"]])
X_train_embarked = si_embarked.fit_transform(X_train[["Embarked"]])

X_test_age = si_age.fit_transform(X_test[["Age"]])
X_test_embarked = si_embarked.fit_transform(X_test[["Embarked"]])                        
```

```python
X_train_age
```

# One Hot Encoding

```python
ohe_sex = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
ohe_embarked = OneHotEncoder(sparse_output=False, handle_unknown="ignore")

X_train_sex = ohe_sex.fit_transform(X_train[["Sex"]])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.fit_transform(X_test[["Sex"]])
X_test_embarked = ohe_embarked.fit_transform(X_test_embarked)
```

```python
X_train_sex.shape
```

```python
X_test_sex.shape
```

```python
X_train_embarked.shape
```

```python
X_train_drop = X_train.drop(columns=["Age", "Sex", "Embarked"])
X_test_drop = X_test.drop(columns=["Age", "Sex", "Embarked"])
```

```python
X_train_transformed = np.concatenate((X_train_drop, X_train_age, X_train_sex, X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_drop, X_test_age, X_test_sex, X_test_embarked),axis=1)
```

```python
X_train_transformed.shape
```

```python
X_test_transformed.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed, y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
