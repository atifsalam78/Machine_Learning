# Encoded Boolean like values (Yes/No)

```python
bool_cols = ["Other_Loans", "Default_History", "Foreign_Worker",
             "Mobile_Banking_Active", "Auto_Pay_Enabled"]
```

```python
for col in bool_cols:
    if col in bank.columns:
        bank[col] = bank[col].replace({"Yes": 1, "No": 0}).copy()
```

```python
bank.sample(5)
```

# Correlation Matrix

```python
num_cols = bank.select_dtypes(include=np.number).columns
```

```python
corr = bank[num_cols].corr()
```

```python
corr
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Matrix - Numeric Features")
plt.show()
```

# Distribution Of Numeric Columns

```python
import math
num_features = bank.select_dtypes(include=np.number).columns
num_features = [col for col in num_features if col in bank.columns]

n_col = 4
n_row = math.ceil(len(num_features)/n_col)

fig, axes = plt.subplots(n_row, n_col, figsize=(15, n_row*4))
axes = axes.flatten()

for i, col in enumerate(num_features):
    sns.histplot(bank[col], kde=True, ax=axes[i], color="skyblue")
    axes[i].set_title(f"Distribution of {col.replace("_", " ")}", fontsize=10)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("")

# Hide empty subplots
for j in range(i+1,len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# Relationship With Target Variable

```python
target = "Credit_Risk"
```

# Box Plot For Continious Features

```python
import math
num_features = [col for col in num_features if col in bank.columns]

n_cols = 3
n_rows = math.ceil(len(num_features)/n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows*4))
axes = axes.flatten()


for i, col in enumerate(num_features):
        sns.boxplot(x=target, y=col, data=bank, ax=axes[i], orient="v")
        axes[i].set_title(f"{col.replace("_", " ")}")

# Remove unused axes

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
bank.describe()
```

```python
import math

cat_features = bank.select_dtypes(include="object").columns.difference(["Customer_ID", "Credit_Risk"])
cat_features = [col for col in cat_features if col in bank.columns]

n_cols = 4
n_rows = math.ceil(len(cat_features)/n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows*4))
axes = axes.flatten()


for i, col in enumerate(cat_features):
        sns.countplot(x=col, hue=target, data=bank, ax=axes[i])
        axes[i].set_title(f"{col.replace("_", " ")} VS {target.replace("_", " ")}")
        axes[i].tick_params(axis="x", rotation=45)
        

# Remove unused axes

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()

plt.show()
```

# Correlation with Credit Risk (if numeric encoding)

```python
if bank[target].dtype != "object":
    corr_target = corr[target].sort_values(ascending=False)
    print("Top correlations with Credit_Risk:\n", corr_target.head(10))
else:
    print("Credit_Risk is categorical — correlations will be handled via group means or encoding.")
```

# Derived Metrics
### Example: Average Spending-to-Income Ratio by Risk

```python
if "Spending_to_Income_Ratio" in bank.columns:
    ratio_summary = bank.groupby(target)['Spending_to_Income_Ratio'].mean()
    print("\nAverage Spending-to-Income Ratio by Credit Risk:\n", ratio_summary)
```

# Pairplot (key features)

```python
key_features = ['Monthly_Income','Monthly_Spending','Credit_Utilization_Ratio',
             'Loan_to_Income_Ratio','Behavioral_Score',target]
existing = [f for f in key_features if f in bank.columns]
sns.pairplot(bank[existing], hue=target)
plt.show()
```

# Categorical Feature Insights

```python
cat_features = [col for col in cat_features if col in bank.columns]
print(f"\n--- {col} Crosstab ---")
display(pd.crosstab(bank[col], bank[target], normalize="index") * 100)
```
